```python
#catagory wise review rating for clothes its differnt rating and for footwear its
#differnt according to medium 
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
#replaced space with the _ to name readable
df.columns=df.columns.str.replace(' ','_')
#after replace we got the purachse_amount(usd) as we need to rename it to purchase_amount 
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'}) 
```

```python
df.columns
```

```python
#create a column age_group 
labels=["young adult", "adult", "middle-aged", "senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)

    
```

```python
df[["age","age_group"]].head(50)
# df["age_group"].tail(5)
```

```python
# create the column purchase_frequency_days 
#we have the column frequency_of purchase but in the text format it willl make harder for the
#anyalsis so we will convert the text into the numbers which will be the easier one for the anyalsis 
frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90
}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```
